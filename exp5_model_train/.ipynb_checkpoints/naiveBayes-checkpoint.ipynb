```python
import pandas as pd
ris = pd.read_pickle("/home/hermuba/resistanceExp/data/anno_sps_df")
df = pd.read_pickle("/home/hermuba/resistanceExp/EcoliGenomes/cdhitResult/ec0913_df")
```

```python
train_drug = ['meropenem', 'gentamicin', 'ciprofloxacin', 'trimethoprim/sulfamethoxaole', 'ampicillin', 'cefazolin']
#those drugs have nearly 50 data

from sklearn.naive_bayes import BernoulliNB
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score                                                      
# a function to chooses and join dataframe
def join_df(drug):
    ris_need = ris.loc[ris['Antibiotic'] == drug][['Genome ID', 'Resistant Phenotype']] # get that drug's data
    df_ris = pd.merge(df, ris_need, left_index = True, right_on = "Genome ID")
    df_ris = df_ris.reset_index()
    df_ris = df_ris.drop(['index'], axis = 1)
    
    # drop na
    d = df_ris.loc[df_ris['Resistant Phenotype'].notnull()]
    #print(d['Resistant Phenotype'])
    
    
    return(d)

# throw into naive bayes
def train_nb(df_ris):
    # hold out test set
    X_train, X_test, y_train, y_test = train_test_split(df_ris.iloc[:, 0:15949], df_ris['Resistant Phenotype'], test_size=0.4, random_state=0)
    
    X_train.reset_index(inplace = True)
    y_train.reset_index(inplace = True, drop = True)
    
    # seperate train and test into k fold
    cv = ShuffleSplit(n_splits=3, test_size=0.3, random_state=0)
    #cv =StratifiedKFold(n_splits=3)
    # bnb.fit
    bnb = BernoulliNB()
    for train_index, test_index in cv.split(X_train):
        #print(y_train)
        #print(y_train[train_index]) # NaN comes from no where!
        bnb.fit(X_train.iloc[train_index,:], y_train[train_index])
        
        val = bnb.predict(X_train.iloc[test_index])
        
        score = accuracy_score(y_train[test_index], val)
        print('accuracy is ', score)
    #cross_val_score(bnb, X_train, y_train, cv=cv) # cross_val_score calls fit but does not return trained estimator
    
    # model.predict()
    bnb.predict(X_train)
```

```python
train_nb(join_df("meropenem"))
```

```python
train_nb(join_df("gentamicin"))
```
